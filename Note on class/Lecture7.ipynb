```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
x = np.linspace(-10,10,1000)
plt.plot(x,np.sin(x), label='Sin')
plt.plot(x,np.cos(x), label='Cos')
# plt.plot(x,np.tan(x), label='Tan')
plt.title('Sin Function')
plt.xlabel('x value')
plt.ylabel('sin value')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14,8))
ax = plt.subplot(2, 2, 1)
plt.plot(x,np.sin(x))
# plt.xlabel('Sin x')

plt.subplot(2, 2, 2)
plt.plot(x,np.cos(x))
# ax.set_xlabel('Sin x')
# ax.set_ylabel('Sin value')
ax.set(xlabel='Sin x',ylabel='Sin Value', title='Axes 1')
plt.xlabel('Cos x')

plt.subplot(2, 2, 3)
plt.plot(x,np.tan(x))
plt.xlabel('Tan x')

plt.tight_layout()
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
fig, ax = plt.subplots(1,2,sharey=True)

fig.set_figheight(5)
fig.set_figwidth(17)

ax[0].scatter(iris.sepal_length, iris.sepal_width, label='Sepal', c='red', marker='*', s=50)
ax[0].scatter(iris.petal_length, iris.petal_width, label='Petal')
ax[0].legend()
ax[0].set(xlabel='X Label', ylabel='Y Label', title='Title')

ax[1].hist(iris.sepal_length, bins=50, label='Sepal Length Distribution')
ax[1].hist(iris.petal_length, bins=50, label='Petal Length Distribution', alpha=0.4)
ax[1].legend()
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.groupby('day').total_bill.sum()
```

```python
plt.bar(tips.groupby('day').total_bill.sum().index,tips.groupby('day').total_bill.sum().values)
plt.show()
```

```python
ax = plt.axes()
ax.bar(tips.groupby('day').total_bill.sum().index,tips.groupby('day').total_bill.sum().values)
plt.show()
```

```python
plt.barh(tips.groupby('day').total_bill.sum().index,tips.groupby('day').total_bill.sum().values)
plt.show()
```

```python
exam = tips.groupby('day').total_bill.sum()
plt.pie(exam.values, explode=[0,0.2,0,0], labels=exam.index, autopct='%0.2f%%', shadow=True,radius=2)
plt.show()
```

```python
exam = tips.groupby('day').total_bill.sum()
plt.pie(exam.values, explode=[0,0.2,0,0], labels=exam.index, autopct='%0.2f%%', shadow=True,radius=2)


my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```
