```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

- Continous variable: uncountable in a range, ex: height,weight, temperature, pressure --> measurable
- Discrete variable: countable in a range, ex: day of weeks, rolling dice --> countable

```python
plt.scatter(tips.tip,tips.total_bill);
```

```python
sns.scatterplot(data=tips, x='tip', y='total_bill', hue='time');
```

```python
sns.histplot(data=tips, x='total_bill',bins=20);
# đếm số lần xuất hiện của 1 khoảng giá trị, biến liên tục
```

```python
sns.barplot(data=tips.groupby('sex').total_bill.mean().to_frame().reset_index(), x='sex', y='total_bill');
#vẽ biến rời rạc
```

### FacetGrid

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips, row='sex')
g.map(sns.scatterplot, 'tip', 'total_bill');
```

```python
g = sns.FacetGrid(data=tips, row='sex', col='smoker')
g.map(sns.scatterplot, 'tip', 'total_bill');
```

```python
g = sns.FacetGrid(data=tips, row='sex', col='smoker')
g.map_dataframe(sns.stripplot, y='tip');
```

```python
g = sns.FacetGrid(data=tips, row='sex', col='smoker')
g.map_dataframe(sns.stripplot, x='day', y='tip');
```

### Catplot

```python
sns.stripplot(data=tips,x='sex',y='tip',hue='smoker');
```

```python
sns.countplot(data=tips,x='sex');
```

```python
sns.catplot(data=tips, x='sex', kind='count');
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='strip');
```

```python
sns.catplot(data=tips, row='sex', col='smoker',hue='time', x='day', y='tip', kind='strip');
```

```python
sns.countplot(data=tips,y='smoker');
```

```python
sns.catplot(data=tips, row='time', col='smoker', x='sex', y='total_bill', kind='box');
```

```python
sns.catplot(data=tips, row='smoker', col='time', x='sex', y='total_bill', kind='violin');
```

### Pair Plot

```python
sns.pairplot(data=tips.iloc[:,:-1],diag_kind='kde');
```

```python
sns.regplot(data=tips, x='tip', y='total_bill',ci=100);
```

- confidence interval = 1 - alpha = 99%, 95%, 90%
- significant level: alpha = 0.01, 0.05, 0.1 --> p_value < alpha
