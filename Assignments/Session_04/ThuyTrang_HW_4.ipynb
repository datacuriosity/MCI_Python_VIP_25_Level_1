# 10- Compare the distinct ads per hour in September, October, and November? Which month and which hour has top ads?

```python
# Lấy mốc thời gian
df['hour'] = pd.DatetimeIndex(df['timestamp']).hour
df['month'] = pd.DatetimeIndex(df['timestamp']).month
df['year'] = pd.DatetimeIndex(df['timestamp']).year
df.head()
```

```python
# Tạo 1 query để chỉ lấy những tháng 9, 10, 11
df_91011 = df.query("month >=9 & month <= 11")
df_91011
```

```python
cate_key_91011 = list(df_91011.category_name.unique())
len(cate_key_91011)
```

```python
# Tạo 1 list rỗng để đưa những tháng có lượng AD cao nhất trong tháng 9, 10, 11 của mỗi item vào list
highest_month_91011 = []
highest_month_91011
```

```python
# Tạo vòng lặp để đưa giá trị vào list
for i in cate_key_91011:
    highest_month_91011.append(df_91011[df_91011.category_name == i].month.value_counts(ascending = False).index[0])
len(highest_month_91011)
```

```python
# Chuyển đổi từ list sang Series
pd_cate_key_91011 = pd.Series(cate_key_91011)
pd_highest_month_91011 = pd.Series(highest_month_91011)
```

```python
type(pd_cate_key_91011)
```

```python
type(pd_highest_month_91011)
```

```python
# Nối 2 Series bao gồm Item quảng cáo và tháng có lượng quảng cáo nhiều nhất của mỗi Item
new_df_91011_month = pd.concat([pd_cate_key_91011,pd_highest_month_91011], axis = 1)
new_df_91011_month.columns = ['Item', 'Month_Highest_number_ad_in_91011']
new_df_91011_month
```

```python
# Tạo 1 list rỗng để đưa những tháng có lượng AD cao nhất trong tháng 9, 10, 11 của mỗi item vào list
highest_hour_91011 = []
```

```python
# Tạo vòng lặp để đưa giá trị vào list
for i in cate_key_91011:
    highest_hour_91011.append(df_91011[df_91011.category_name == i].hour.value_counts(ascending = False).index[0])
len(highest_hour_91011)
```

```python
# Chuyển đổi từ list sang Series
pd_highest_hour_91011 = pd.Series(highest_hour_91011)
type(pd_highest_hour_91011)
```

```python
# Nối 2 Series bao gồm Item quảng cáo và tháng có lượng quảng cáo nhiều nhất của mỗi Item
new_df_91011_hour = pd.concat([pd_cate_key_91011,pd_highest_hour_91011], axis = 1)
new_df_91011_hour.columns = ['Item', 'hour_Highest_number_ad_in_91011']
new_df_91011_hour
```

```python
#Concat cả 2 Dataframe thành 1 bảng tổng thể:
new_df_month_hour = pd.concat([new_df_91011_month,pd_highest_hour_91011], axis = 1, join = 'inner')
new_df_month_hour.columns = ['Item', 'Month_Highest_number_ad_in_Sep&Oct&Nov', 'Hour_Highest_number_ad_in_Sep&Oct&Nov']
```
