# Seaborn Exercises

## The Data

We will be working with a famous titanic data set for these exercises.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

## plot join distribution of 'fare' as x and 'age' as y using ``seaborn.jointplot``

```python
sns.jointplot(data=titanic, x="fare", y="age");
```

## plot distribution of 'fare' using ``seaborn.distplot``

```python
sns.displot(titanic['fare']);
```

## plot distribution of 'age' with 3 difference 'class' using ``seaborn.boxplot``

```python
sns.boxplot(y='age',x='pclass',data=titanic);
```

## same as above, but this time let's use ``seaborn.swarmplot``

```python
sns.swarmplot(data=titanic, x='pclass',y='age');
```

## Use ``seaborn.countplot`` to show number of 'male' and 'female'

```python
sns.countplot(data=titanic,x='sex');
```

## Use ``pandas.corr`` to show correlation metrix for data fields

```python
import pandas as pd
titanic.corr()
```

## Then plot correlation matrix using ``seaborn.heatmap``

```python
sns.heatmap(titanic.corr(),cmap="coolwarm");
```

## Use ``seaborn.FacetGrid`` mapped with ``plt.hist`` to plot histogram of ``age`` for 'male' & 'female'

```python
g = sns.FacetGrid(data=titanic,col='sex');
g.map(sns.histplot,'age');
```
