```python
import pandas as pd
import numpy as np
df2 = pd.read_csv('C:/Users/hp/Documents/GitHub/MCI_Python_VIP_25_Level_1/sample_data.csv')
```

```python
df2
```

```python
# 1- How many distinct account_id on each platform?
df2.groupby('platform')[['account_id']].agg(['count', 'nunique']) #nunique trả về các giá trị không trùng nhau
```

```python
# 2- How many distinct ad_id on each platform?
```

```python
df2.groupby('platform')[['ad_id']].agg(['count', 'nunique'])
```

```python
#c2:
df2.groupby('platform').ad_id.nunique()
```

```python
# 3- For each category_name how many ad is advertised?
df2.groupby('category_name').ad_id.count()
```

```python
# 4- Give me top 5 category_name having the highest number of ads?
df2.groupby('category_name').ad_id.count().sort_values(ascending=False).head(5)
```

```python
# 5- Give me top 5 category_name having lowest number of ads?
df2.groupby('category_name').ad_id.count().sort_values(ascending = False).tail(5)
```

```python
# 6- Give me the average price for all distinct ads within each category?
df2.groupby('category_name').price.mean()
```

```python
# 7- Give me the average price for all distinct ads within each region_name?
df2.groupby('region_name').price.mean()
```

```python
# 8- Give details of condition_ad_name for all ads within each category? For example, for category cars, 
#how may ads is 'Moi', 'Da su dung', and 'NaN'
df2.head(10)
df2.pivot_table(values = 'ad_id', index = 'category_name', columns = 'condition_ad_name', aggfunc = 'count')
```

```python
# 9 - Which month in 2018 we see the highest number of (distinct ads)?
df3 = df2['timestamp']
a = np.asarray(df3)
b = []
for i in a:
    b.append(i[2:7])
df2.insert(2, column = 'Date', value = b)
df2.groupby('Date').ad_id.count().sort_values(ascending = False).head(1)
```

```python
# 10- Compare the distinct ads per hour in September, October, and November? Which month and 
#which hour has top ads?
b = []
c = []
for i in a:
    b.append(i[5:7] + i[11:13])
    if i[5:7] in ('09', '10', '11'):
        c.append(1)
    else:
        c.append(0)
df2.insert(2, column = 'Datetime', value = b)
df2.insert(2, column = "Sub1", value = c)
df3 = df2.query('Sub1 == 1')
df3.groupby('Datetime').ad_id.count().sort_values(ascending = False)

#if i[5:7] in ('09', '10', '11'):
```
