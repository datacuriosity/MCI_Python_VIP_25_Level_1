```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- Figure
- Axes

```python
plt.figure(figsize=(10, 6))
x = np.linspace(-10, 10, 1000)
plt.plot(x, np.sin(x), label='Sin function')
plt.plot(x, np.cos(x), label='Cos function')
plt.title('Sin function')
plt.xlabel('x values')
plt.ylabel('sin values')
plt.legend(loc=1)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 5))
ax = plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
# plt.xlabel('Sin x label')

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.xlabel('Cos x label')

ax.set(xlabel='Sin', ylabel='Values', title='Axes 1')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 10))
ax = plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
# plt.xlabel('Sin x label')

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.xlabel('Cos x label')

ax.set(xlabel='Sin', ylabel='Values', title='Axes 1')
plt.tight_layout()
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head(2)
```

```python
iris.species.unique()
```

```python
fig, axes = plt.subplots(1, 2, sharey=True)
fig.set_figwidth(15)

axes[0].scatter(iris.sepal_length, iris.sepal_width, label='Sepal', marker='*', s=200)
axes[0].scatter(iris.petal_length, iris.petal_width, label='Petal')
axes[0].legend()
axes[0].set(xlabel='X label', ylabel='Y label', title='Title')

axes[1].hist(iris.sepal_length, bins=50, label='Sepal lengh distribution')
axes[1].hist(iris.petal_length, bins=50, label='Petal lengh distribution', alpha=0.8)
axes[1].legend()
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(2)
```

```python
tips.groupby('day').total_bill.sum()
```

```python
plt.bar(tips.groupby('day').total_bill.sum().index, tips.groupby('day').total_bill.sum().values)
plt.show()
```

```python
axes = plt.axes()
axes.bar(tips.groupby('day').total_bill.sum().index, tips.groupby('day').total_bill.sum().values)
plt.show()
```

```python
plt.barh(tips.groupby('day').total_bill.sum().index, tips.groupby('day').total_bill.sum().values)
plt.show()
```

```python
example = tips.groupby('day').total_bill.sum()
plt.pie(example.values, explode=[0, 0.2, 0, 0], labels=example.index, autopct='%0.2f%%', shadow=True, radius=2)
plt.show()
```

```python
plt.pie(example.values, explode=[0, 0.2, 0, 0], labels=example.index, autopct='%0.2f%%', shadow=True, radius=2)

# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```
