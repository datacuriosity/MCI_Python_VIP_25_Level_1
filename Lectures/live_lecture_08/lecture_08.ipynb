```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

- Continuous variable: heigh, weight, temperature, pressure --> measurable
- Discrete variable: rolling dice, days of weeks: countable

```python
plt.scatter(tips.tip, tips.total_bill);
```

```python
sns.scatterplot(data=tips, x='tip', y='total_bill', hue='time');
```

- Bar chart
- Histogram

```python
sns.histplot(data=tips, x='total_bill', bins=20);
```

```python
tips.groupby('sex').total_bill.mean().to_frame().reset_index()
```

```python
sns.barplot(data=tips.groupby('sex').total_bill.mean().to_frame().reset_index(), x='sex', y='total_bill');
```

```python
tips.groupby('sex').total_bill.mean().plot(kind='bar');
```

### FacetGrid

```python
tips.head(2)
```

```python
g = sns.FacetGrid(data=tips, row='sex')
g.map(sns.scatterplot, 'tip', 'total_bill');
```

```python
g = sns.FacetGrid(data=tips, row='sex', col='smoker')
g.map(sns.scatterplot, 'tip', 'total_bill');
```

```python
g = sns.FacetGrid(data=tips, row='sex', col='smoker')
g.map_dataframe(sns.stripplot, x='day', y='tip')
plt.legend();
```

### Catplot

```python
sns.stripplot(data=tips, x='sex', y='tip', hue='smoker');
```

```python
sns.countplot(data=tips, x='sex');
```

```python
sns.catplot(data=tips, x='sex', kind='count');
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='strip');
```

```python
sns.catplot(data=tips, col='smoker', x='sex', y='total_bill', kind='strip');
```

```python
sns.catplot(data=tips, col='time', row='smoker',  x='sex', y='total_bill', kind='strip');
```

```python
sns.catplot(data=tips, col='time', row='smoker',  x='sex', y='total_bill', kind='violin');
```

```python
sns.catplot(data=tips, col='smoker', row='time',  x='sex', y='total_bill', kind='box');
```

### pair plot

```python
sns.pairplot(data=tips.iloc[:, :-1], diag_kind='kde');
```

```python
sns.regplot(data=tips, x='tip', y='total_bill', ci=99);
```

- confidence interval = 1- alpha = 99%, 95%, 90%
- significant level: alpha = 0.01, 0.05, 0.1  --> p_value < alpha
