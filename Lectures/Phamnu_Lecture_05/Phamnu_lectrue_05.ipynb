### Common Function

#### Group by: gom nhóm
##### Agrregation function: count, mean, max, min,...

```python
df.groupby('sex').total_bill.sum().to_frame().rename(columns = {'total_bill': 'sum_total_bill'}) # lấy tổng số tiền total bill của mỗi giới tính, chuyển về dataframe vì kiểu trả về là series
#rename: đổi tên cột
```

```python
df.groupby('sex')[['total_bill', 'tip']].agg(['mean', 'max', 'min', 'sum'])
```

```python
df.groupby('sex').agg({"total_bill": ['mean', 'max'], "tip": ["min", "sum"]}) # tính giá trị của mỗi đối tượng khác nhau
```

```python
# đối với sex = female và smoke = no => tổng bill là bao nhiêu
df.pivot_table(values = 'total_bill', index ='sex', columns = 'smoker', aggfunc = 'sum')
```
