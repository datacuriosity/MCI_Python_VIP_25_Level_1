### Indexing

- iloc[row_index, column_index]
- loc[row_name, column_name]

```python
indices = [f'index_{i}' for i in range(10)]
df_10 = df.head(10)
df_10.index = indices
df_10
```

```python
df_10.iloc[2: 5, :]
```

```python
df_10.loc['index_2': 'index_4', :]
```

```python
df_10.iloc[2: 5, 1:3]
```

```python
df_10.loc['index_2': 'index_4', 'tip': 'sex']
```

```python
df_10
```

```python
df_10.iloc[[1, 3, 5]]
```

```python
df_10.loc[['index_1', 'index_3', 'index_5']]
```

```python
df_10.iloc[::2]
```

### Common Functions

Groupby: aggregation function: count, mean, max, min, sdt, ...

```python
df.groupby('sex').total_bill.sum().to_frame().rename(columns={'total_bill': 'sum_total_bill'})
```

```python
df.groupby('sex')[['total_bill', 'tip']].agg(['mean', 'max', 'sum', 'min'])
```

```python
df.groupby('sex').agg({'total_bill': ['mean', 'sum'], 'tip': ['max', 'min']})
```

```python
df
```

```python
df.pivot_table(values='total_bill', index='sex', columns='smoker', aggfunc='sum')
```
